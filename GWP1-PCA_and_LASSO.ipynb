## 1. Financial Products

#### 1.1 What is an Exchange-Traded Fund (ETF)?

Exchange-Traded Funds (ETFs) are mutual funds that invest in stocks based on a theme, sector, or index. After that, the corporation sells these funds' shares and collects an annual expense fee.

ETFs vs Mutual Funds
- ETFs are less expensive, average annual fees to hold an ETF are 0.4% versus fees of 1% and more for mutual funds
- ETFs trade like stocks, you can buy or sell them throughout the day
- ETFs only create a tax liability when you sell

Benefits of ETF Investing
- Easy exposure to an entire investing theme, sector or group of investments
- Stress-free strategy with your money spread across hundreds of stocks and other assets
- Hybrid investing strategy for diversification AND returns

Drawbacks to ETF Investing
- Too much diversification makes it tough to beat the market!
- High expense ratios on some funds
- Overly complicated and risky strategies in some funds

How to Pick an ETF
- Groups of asset classes or sectors to fill gaps in your portfolio
- Investing themes that might do well given big picture economic factors or trends

ETF Groups to Fill Gaps in your Portfolio
- Asset classes like stocks, bonds, real estate and alternative investments
- Sectors like utilities, consumer goods, energy or industrials
- Company size like small-cap or large-cap stocks

How to Compare ETFs
- Expense Ratio
- Dividend Yields and Total Returns
- Portfolio Holdings

## 3. 1-Dimensional Data Summaries

#### 3.1 Summarize the min, max, mean, median, and standard deviation of each column

```python
def data_summaries(df):
    """
    Returns the summary statistic of all columns in the given DataFrame in the order below:
    min, max, mean, median, standard deviation
    """
    
    summary = df.describe().drop(['count', '25%', '75%'], axis=0)
    summary_reindexed = summary.reindex(['min', 'max', 'mean', '50%', 'std']).rename({'50%': 'median'}, axis=0)

    return summary_reindexed
```

## 4. Data Visualizing

#### 4.1 Write a Python function that graphs 1 time series with appropriate time labels

```python
def one_time_series_graph(data, etf_name):
    """
    This function graphs 1 time series with appropriate time labels:
    data = DataFrame containing all ETFs
    etf_name = Name of the ETF as string from the DataFrame you wish to graph it's time series
    """
    
    plt.figure(figsize=(12, 6))
    plt.plot(pd.to_datetime(data['Date']), data[etf_name])
    plt.grid(which='both', axis='both')
    plt.title('Line graph showing the {} ETF from {} to {}\n'.format(etf_name,
                                                                     str(pd.to_datetime(data['Date'])[0]).split(' ')[0],
                                                                     str(pd.to_datetime(data['Date'])[len(etf)-1]).split(' ')[0]
                                                                    ))
    plt.legend([etf_name])
    plt.xlabel('Time in years')
    plt.ylabel('{} ETF'.format(etf_name))
    plt.show()
```

```python
one_time_series_graph(etf, 'MSCI USA')
```

#### 4.2 Write a Python function that graphs 2 time series on the same plot, with labels

```python
def two_time_series_graph(data, etf_name_1, etf_name_2):
    """
    This function graphs 1 time series with appropriate time labels:
    data = DataFrame containing all ETFs
    etf_name_1 = Name of the first ETF as string from the DataFrame you wish to graph it's time series
    etf_name_2 = Name of the second ETF as string from the DataFrame you wish to graph it's time series
    """
    
    plt.figure(figsize=(12, 6))
    plt.plot(pd.to_datetime(data['Date']), data[[etf_name_1, etf_name_2]])
    plt.grid(which='both', axis='both')
    plt.title('Line graph showing {} and {} ETFs from {} to {}\n'.format(etf_name_1, etf_name_2,
                                                                     str(pd.to_datetime(data['Date'])[0]).split(' ')[0],
                                                                     str(pd.to_datetime(data['Date'])[len(etf)-1]).split(' ')[0]
                                                                    ))
    plt.legend([etf_name_1, etf_name_2])
    plt.xlabel('Time in years')
    plt.ylabel('ETF returns')
    plt.show()
```

```python
two_time_series_graph(etf, 'MSCI UK', 'MSCI USA')
```

#### 4.3 Write a Python function that compares the 2 return series

```python
def compare_two_returns(data, etf_name_1, etf_name_2):
    """
    This function graphs 1 time series with appropriate time labels:
    data = DataFrame containing all ETFs
    etf_name_1 = Name of the first ETF as string from the DataFrame you wish to graph it's time series
    etf_name_2 = Name of the second ETF as string from the DataFrame you wish to graph it's time series
    """
       
    plt.figure(figsize=(12, 6))
    plt.plot(pd.to_datetime(data['Date']), data[[etf_name_1, etf_name_2]].pct_change())
    plt.grid(which='both', axis='both')
    plt.title('Line graph showing the Percentage change of {} and {} ETFs from {} to {}\n'.format(etf_name_1, etf_name_2,
                                                                     str(pd.to_datetime(data['Date'])[0]).split(' ')[0],
                                                                     str(pd.to_datetime(data['Date'])[len(etf)-1]).split(' ')[0]
                                                                    ))
    plt.legend(['pct_change ' + etf_name_1, 'pct_change ' + etf_name_2])
    plt.xlabel('Time in years')
    plt.ylabel('ETF returns')
    plt.show()
```

```python
compare_two_returns(etf, 'MSCI UK', 'MSCI USA')
```

## 7. Principal Component Analysis

#### 7.1 Use the Pearson correlation matrix

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Scale the data
X_scaled = StandardScaler().fit_transform(X)

# Compute the pca setting the n_components as 10, fit and transform the scaled datap
pca = PCA(n_components=10)
pca_X = pca.fit_transform(X_scaled)

pca_X
```

#### 7.2 Show the amount of variation explained by the first 5 components

```python
pca_explained_variance_ratio = pca.explained_variance_ratio_
first_5_components = np.sum(pca_explained_variance_ratio[0:5])

print('The amount of varation explained by the first 5 components is: {:.4%}'.format(first_5_components))
x_axis = ['PC'+ str(i) for i in range(1,11)]

plt.figure(figsize=(12, 6))
plt.plot(x_axis, pca_explained_variance_ratio)
plt.title('Graph showing the pca explained variance ratio\n')
plt.xlabel('Principal Components')
plt.ylabel('Explained Variation Ratio')
plt.show()
```

#### 7.3 How many components are needed to express 80% of the variation of the data?

```python
first_2_components = np.sum(pca_explained_variance_ratio[0:2])

print('The first two components are sufficient to express 80% of the variation of the data: {:.2%}'
      .format(first_2_components))
```

#### 7.4 What is your interpretation of the 1st component?

```python
PC1 = pca_explained_variance_ratio[0]

print('The first component (PC1) explains {:.2%}'.format(PC1), 
      'of the variation, This implies that the line of best fit is best approximated by \nthe first component.')
```

#### 10.3 Split the data into testing and training, with 2/3 for training and 1/3 for testing

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.3)
```

#### 10.2 For the lasso, use at least 1000 different values of the penalty parameter

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
```

```python
mse = []
reg_list = []

for params in np.arange(0.1, 10.1, 0.01):
    lasso_regressor = Lasso(alpha = params)
    lasso_regressor.fit(X_train, y_train)
    y_lassopred = lasso_regressor.predict(X_test)
    mismatch = mean_squared_error(y_lassopred, y_test) 
    reg_list.append(params)
    mse.append(mismatch)
```

#### 10.4 Graph the overall model mismatch for each of the 1000 values of the paramter

```python
plt.figure(figsize=(12, 6))
plt.plot(reg_list, mse)
plt.title('Graph showing the overall model mismatch for each of the 1000 values of the parameter\n')
plt.xlabel('Parameter value')
plt.ylabel('Mismatch: mean squared error')
plt.grid(which='both', axis='both')
plt.show()
```

#### 10.5 Find a lasso model that includes no more than 7 predictors

```python
reg_list2 = []
coeff_count = []

for params in np.arange(0.1, 10.1, 0.01):
    lasso_regressor = Lasso(alpha = params)
    lasso_regressor.fit(X_train, y_train)
    coeff1 = lasso_regressor.coef_
    y_lassopred = lasso_regressor.predict(X_test)
    count = (np.abs(coeff1)>0.1).sum()
    reg_list2.append(params)
    coeff_count.append(count)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(reg_list2, coeff_count)
plt.title('Graph showing number of predictors with "Large" co-efficients \n')
plt.xlabel('r_value')
plt.ylabel('Coefficient count')
plt.grid(which='both', axis='both')
plt.show()
```

```python
# Now we choose our value of r = 8 based on the diagram above

r = 8
lasso_regressor = Lasso(alpha = r)
lasso_regressor.fit(X_train, y_train)
coeff = lasso_regressor.coef_
count = (np.abs(coeff)>0.1).sum()

# Next, we choose our 7 independent variables that have highest co-efficients
chosen_variables = np.argsort(np.abs(coeff))[-7:]
X_train.columns[chosen_variables]
```
